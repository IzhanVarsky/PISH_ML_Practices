# Правила игры и система оценивания

## 1. Домашние задания
- Всего **4 домашних задания**.  
- Каждое задание оценивается **до 25 баллов**
- Баллы могут снижаться за ошибки, неточности и недостаточные ответы на мои теоретические вопросы.  
- Максимум за домашние задания — **100 баллов**
- Задание сдано, если за него получено более 12.5 баллов

### Необязательное задание
- Одно из четырёх домашних заданий (по выбору) является **необязательным**, т.е. важно сдать хотя бы 3 задания для получения **допуска** к Хакатону
- Если студент его не сдаёт, получает за него 0 баллов; если сдаёт — баллы учитываются как обычно

---

## 2. Хакатон
В конце курса проводится хакатон. Студенты делятся на **3 команды**.

### Результаты:
- **Не пришёл/не допущен**: −20 баллов  
- **командное 3-е место**: каждому −10 баллов  
- **командное 2-е место**: каждому +10 баллов  
- **командное 1-е место**: каждому +20 баллов  

---

## 3. Финальный балл и оценка
Сумма домашних заданий + баллы за хакатон формируют итоговый результат.

| Итоговый балл | Оценка |
|---------------|--------|
| < 60          | не сдал |
| 60–75         | 3 |
| 75–90         | 4 |
| > 90          | 5 |

#### Подробности об оценке домашних заданий

- Штрафов за подходы или за ошибки не будет (но просьба систему не абьюзить)
- Оценка за дз будет равна проценту того, что вы выполнили относительно требуемого по заданию. Например, если сделали половину работы, то я ставлю 50%. Затем можно опять подойти и досдать и если доделали до 80%, то поставлю 80%
- При этом если плохо отвечаете на мои вопросы (в том числе плохо понимаете свой код), то буду снижать
- Из штрафов есть только штраф за дедлайн, который нужен, чтобы мотивировать вас не откладывать всё на конец курса. Работает так: чем раньше вы подойдете ко мне (желательно на следующий учебный день после дня выдачи дз) и покажете, что вы уже начали делать дз и у вас есть какие-то более-менее средние успехи (т.е. не обязательно что всё сразу сделано), то я фиксирую дедлайн, и потом вы можете еще подходить и досдавать уже без штрафа. 
Пример: если 1 лабу вы показываете мне завтра (несмотря на то что сделали её не полностью), то штрафа за дедлайн не будет, мы пообщаемся, я как-нибудь что-нибудь подскажу, и вы сможете потом еще досдавать эту лабу мне без штрафов. Если же вы подойдете ко мне первый раз по первой лабе в субботу, то будет штраф 12%. Если первый раз подойдете в следующий вт, то штраф 24% и т.п.
