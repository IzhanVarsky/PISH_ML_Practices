|-----------------------------------------------|---------|
| **–ö–∞—á–µ—Å—Ç–≤–æ –º–æ–¥–µ–ª–∏ (MAE, RMSE, R¬≤, MAPE)**     | **50%** |
| –ö–∞—á–µ—Å—Ç–≤–æ –¥–∞–Ω–Ω—ã—Ö (–ø–∞—Ä—Å–∏–Ω–≥, —Å—Ç—Ä—É–∫—Ç—É—Ä–∞, –æ—á–∏—Å—Ç–∫–∞) | 15%     |
| –£–¥–æ–±—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Å–µ—Ä–≤–∏—Å–æ–º                 | 15%     |
| Feature Engineering –∏ EDA                     | 10%     |

## –ö–æ–¥ –¥–ª—è –æ—Ü–µ–Ω–∫–∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –≤–æ –≤—Ä–µ–º—è –ø—Ä–µ–∑–µ–Ω—Ç–∞—Ü–∏–∏

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# –î–∞–Ω–Ω—ã–µ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ (–∑–∞–º–µ–Ω–∏—Ç–µ –Ω–∞ —Ä–µ–∞–ª—å–Ω—ã–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –≤–∞—à–∏—Ö –º–æ–¥–µ–ª–µ–π)
real_prices = [75000, 120000, 45000, 89000, 156000, 300000, 22000]  # —Ä–µ–∞–ª—å–Ω—ã–µ —Ü–µ–Ω—ã

predicted_prices = [82000, 115000, 50000, 85000, 145000, 280000, 30000]  # –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫
def calculate_metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    mape = np.mean(np.abs((np.array(y_true) - np.array(y_pred)) / np.array(y_true))) * 100
    
    return {
        'MAE': round(mae, 2),
        'RMSE': round(rmse, 2), 
        'R¬≤': round(r2, 4),
        'MAPE': round(mape, 2)
    }

# –°—á–∏—Ç–∞–µ–º –º–µ—Ç—Ä–∏–∫–∏
metrics = calculate_metrics(real_prices, predicted_prices)

# –ö—Ä–∞—Å–∏–≤–æ–µ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ
print("üìä –†–ï–ó–£–õ–¨–¢–ê–¢–´ –ü–†–ï–î–°–ö–ê–ó–ê–ù–ò–ô")
print("=" * 40)
for metric, value in metrics.items():
    if metric == 'MAPE':
        print(f"{metric}: {value}%")
    elif metric == 'R¬≤':
        print(f"{metric}: {value}")
    else:
        print(f"{metric}: {value} ‚ÇΩ")

print("\nüìà –î–ï–¢–ê–õ–ò –ü–†–ï–î–°–ö–ê–ó–ê–ù–ò–ô:")
print("–†–µ–∞–ª—å–Ω–∞—è —Ü–µ–Ω–∞ | –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ  | –û—à–∏–±–∫–∞")
print("-" * 45)
for true, pred in zip(real_prices, predicted_prices):
    error = pred - true
    error_percent = (error / true) * 100
    print(f"{true:>11} ‚ÇΩ | {pred:>11} ‚ÇΩ | {error:>+7.0f} ‚ÇΩ ({error_percent:>+6.1f}%)")
```
