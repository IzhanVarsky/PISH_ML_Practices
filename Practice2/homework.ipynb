# Домашнее задание №2: Логистическая регрессия (градиентный спуск и sklearn, многоклассовая классификация)

Цель задания — реализовать многоклассовую логистическую регрессию через градиентный спуск, сравнить её с моделью из sklearn и исследовать влияние структуры данных на качество классификации.

---

## 1. Генерация двух датасетов

Создайте два многоклассовых датасета с тремя классами и двумя признаками.

### Датасет A (простой)
- классы хорошо разделимы визуально;
- примерно равное количество объектов каждого класса;
- не менее 1000 объектов.

### Датасет B (сложный)
- классы частично перекрываются;
- классы несбалансированы (например, пропорции около 0.7 / 0.2 / 0.1);
- тоже не менее 1000 объектов.

Для каждого датасета:
1. Разделите данные на train/test (стратифицированно!).
2. Выполните нормализацию признаков.
3. Постройте 2D scatter-график точек с раскраской по классам.

---

## 2. Softmax-регрессия через градиентный спуск (для обоих датасетов)

Реализуйте многоклассовую логистическую регрессию вручную через градиентный спуск:

- прямой проход (логиты → softmax → вероятности),
- вычисление функции потерь,
- вычисление градиента,
- обновление параметров на каждой эпохе.

Выберите один learning rate, обеспечивающий устойчивую сходимость, и обучите модель.

### Что нужно сделать:
1. График убывания функции потерь (train и test).
2. График изменения accuracy и других метрик по эпохам (train и test).
3. Визуализацию решающих областей (decision regions) на train и test.
4. Показать финальные значения для обоих выборок:
   - accuracy,
   - macro F1,
   - weighted F1,
   - confusion matrix.

---

## 3. Многоклассовая логистическая регрессия через sklearn (для обоих датасетов)

Для каждого датасета A и B обучите логистическую регрессию из sklearn.

### Для каждого датасета:
1. Обучите модель на train.
2. Оцените на train и test:
   - accuracy,
   - macro F1,
   - weighted F1,
   - confusion matrix.
3. Постройте decision regions на train и test.
4. Выведите классификационный отчёт (classification report) и проанализируйте значения precision/recall/F1 для каждого класса.

---

## 4. Сравнение градиентного спуска и sklearn (на обоих датасетах)

Сравните две модели:
- вашу ручную реализацию softmax-регрессии,
- модель из sklearn.

### Нужно сравнить:
1. Метрики accuracy, macro F1, weighted F1.
2. Confusion matrix.
3. Две визуализации decision regions.

Кратко опишите, совпадают ли границы классов, и в чём могут быть причины различий.

---

## 5. Влияние структуры данных (датасеты A vs B)

Для обоих методов и для обоих датасетов, сравните:

1. Значения метрик.
2. Какие классы чаще всего путаются.
3. Как несбалансированность и перекрытие классов влияют на качество:
   - почему accuracy может оставаться высокой,
   - почему macro F1 падает сильнее,
   - почему weighted F1 иногда выглядит лучше macro F1.

Сделайте краткий вывод (несколько предложений).
