# Домашнее задание: Titanic + kNN и бустинг

Возьмите датасет **Titanic** (файл `train.csv` из соревнования):  
https://www.kaggle.com/competitions/titanic/overview

Загруженный мной файл titanic_train.csv аналогичен файлу с каггла. 

## Задача

Предсказать признак `Survived` (0/1) по информации о пассажире.

## Что нужно сделать

1. **Загрузка и EDA**
   - Загрузите данные в `pandas`.
   - Посмотрите на признаки, типы столбцов, распределения.
   - Разберитесь с пропусками:
     - где есть NaN,
     - как вы их заполняете (для числовых и категориальных признаков).
   - **Разбиение на выборки** - отложите **15%** данных в качестве тестовой выборки (`test_size=0.15`).

2. **Подготовка признаков**
   - Выберите набор признаков для обучения (возможно есть бесполезные).
   - Обработайте категориальные признаки, если нужно
   - Масштабируйте числовые признаки (рекомендуется `StandardScaler` или `MinMaxScaler`), если нужно

3. **Визуализация** для анализа данных
   - Выполните понижение размерности (например, **PCA** до 2 компонент).
   - Нарисуйте точки на плоскости (цветом отметьте выживших / невыживших)

4. **Модели**
   - Обучите хотя бы **одну модель бустинга** для классификации
   - Обучите **kNN-классификатор** (`KNeighborsClassifier`)
   - Для обеих моделей посчитайте **classification report** на тестовой выборке

5. **Оценка качества и выводы**
   - Сравните качество моделей по `precision / recall / f1 / accuracy`.
   - Критерии (ориентировочно):
     - `accuracy > 0.70` — **отлично**,
     - `accuracy > 0.80` — **супер**.
   - Оформите общую таблицу с результатами 

6. Отправка решения в kaggle (по желанию = необязательно)
   - Если желаете проверить корректность своего подхода - обучите свой подход на всем датасете, создайте на основе test.csv финальный .csv с предсказаниями и отправьте на kaggle
   - При наборе более 0.8 очков поставлю +7 баллов за весь курс, от 0.7 до 0.8 поставлю +5 баллов за курс, от 0.6 до 0.7 поставлю +3 балла
